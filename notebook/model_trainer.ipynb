```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold,train_test_split,cross_val_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=20,random_state=42)
```

```python
scalar = StandardScaler()
X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)
import pickle
with open("artifacts/scalar.pkl", "wb") as file:
    pickle.dump(scalar, file)
```

```python
y_train['Good/Bad'].unique()
```

```python
y_train['Good/Bad']=y_train['Good/Bad'].map({-1:0,1:1})
y_test['Good/Bad']=y_test['Good/Bad'].map({-1:0,1:1})
```

```python
model1 = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)
model2 = RandomForestClassifier(
    n_estimators=100,
    criterion="gini",
    max_depth=None,
    random_state=42
)

```

```python
def model_trainer(model):
    # Stratified K-Fold Cross Validation
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    cv_scores = cross_val_score(
        model,
        X_train_scaled,
        y_train,
        cv=skf,
        scoring='accuracy'
    )
    
    print("Cross Validation Results")
    print(f"CV Scores      : {cv_scores}")
    print(f"Mean Accuracy  : {cv_scores.mean():.4f}")
    print(f"Std Deviation  : {cv_scores.std():.4f}")
    print("-" * 40)

    model.fit(X_train_scaled,y_train)
    y_pred = model.predict(X_test)
    from sklearn.metrics import accuracy_score, classification_report
    acc = accuracy_score(y_test,y_pred)
    cls = classification_report(y_test,y_pred)
    print("Model Performance")
    print(f'Accuracy is :- {acc:.4f}')
    print('-'*40)
    print(f'Classsification Report')
    print(cls)
```

```python
model_trainer(model1)
```

```python
model_trainer(model2)
```
